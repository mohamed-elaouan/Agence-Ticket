```python
import numpy as np
import matplotlib.pyplot as plt
#Solution
x=4*np.random.rand(500,1)
y=3*x+2+60*np.random.rand(500,1)
plt.scatter(x,y)
```

```python
#POUR trouver
from sklearn.linear_model import SGDRegressor
#pour separe db part test et part traitement
from sklearn.model_selection import train_test_split
#error quadratique
from sklearn.metrics import mean_squared_error
model = SGDRegressor(max_iter=100,random_state=42)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
x_test.shape
```

```python

errors=[]
params=[]
A=[]
B=[]
for i in range(100):
  #calc params 'partial_fit'
  model.partial_fit(x_train,y_train.flatten())
  y_pred=model.predict(x_test)
  eqm=mean_squared_error(y_test,y_pred)
  errors.append(eqm)
  A.append(model.coef_[0])
  B.append(model.intercept_)
```

```python
plt.plot(errors)
A= model.coef_[0]
b= model.intercept_
plt.scatter(x,y, cmap='summer')
plt.plot(x, A*x+b, lw=3, c='y')
```
